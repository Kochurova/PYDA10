## Функции
Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):

p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;

s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится
Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ.

l – list – команда, которая выведет список всех документов в формате
passport "2207 876234" "Василий Гупкин"

as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. Предусмотрите случай, когда пользователь добавляет полку, которая уже существует.

d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. Предусмотрите сценарий, когда пользователь вводит несуществующий документ;

m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку;

a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку.

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def is_doc(_sDoc: str):
    """Возвращает истину если есть документ"""
    lRet = False 
    for i in documents:
        if (i['number'] == _sDoc):
            lRet = True
            break
    return lRet

def is_shelf(_sShelf: str):
    """Возвращает истину если есть полка"""
    lRet = False 
    if _sShelf in directories:
            lRet = True
    return lRet

def get_name(_sDoc: str):
    """Возвращает ФИО человека по введенному номеру документа"""
    fio = 'ФИО не найдено!'
    doc_num = _sDoc
    for i in documents:
        if (i['number'] == doc_num):
            fio = i['name']
            break
    return fio

def get_shelf(_sNumDoc: str): 
    """Возвращает номер полки, на которой лежит документ, по введенному номеру документа"""
    shelf_number = None
    numdoc = _sNumDoc
    print(numdoc)
    for key, value in directories.items():
        for ii in value:
            if (ii == numdoc):
                shelf_number = key 
                break
    return shelf_number

def add_shelf(shnum:str):
    """Добавляет новую полку"""
    shelf_num = shnum
    rets = ''
    if (list(directories.keys()).count(shelf_num) == 0):
        directories[shelf_num] = []
        rets = 'Новая полка:' + shelf_num + ' добавлена'  
    else:
        rets = 'Такая полка уже есть'
    return rets

def get_list():
    for i in documents:
        print (i['type'] + ' "' + i['number'] + '"' + '"' + i['name'] + '"')
    return

def move_doc(dnum: str, shnum:str):
    doc_num = dnum
    shelf_num = shnum # номер целевой полки
    sRet = ''
    cur_key = get_shelf(doc_num)
    if (cur_key != None): # наличие документа на полке
        value = directories.get(cur_key) 
        value.remove(doc_num)
        newvalue = directories.get(shelf_num) 
        newvalue.append(doc_num)
        sRet = 'Документ успешно перемещен'
    else:
        sRet = 'Нет полки с таким документом'
    return sRet

def del_doc(dnum: str):
    doc_num = dnum
    sRet = ''
    cur_key = get_shelf(doc_num)
    if (cur_key != None): # наличие документа на полке
        value = directories.get(cur_key) 
        value.remove(doc_num)
        for i in documents:
            if (i['number'] == doc_num):
                documents.remove(i)
                break
        sRet = 'Документ успешно удален'
    else:
        sRet = 'Нет полки с таким документом'
    return sRet

def add_doc(_doc_num:str, _shelf_num:str, _doc_type:str, _p_name:str):
    doc_num = _doc_num
    shelf_num = _shelf_num
    doc_type = _doc_type
    p_name = _p_name
    sRet = ''
    if (doc_type in ('passport','insurance','invoice','license')):
        documents.append({'type':doc_type, 'number':doc_num, 'name':p_name})
        newvalue = directories.get(shelf_num) 
        newvalue.append(doc_num)
        sRet = 'Документ успешно добавлен'
    else:
        sRet = 'Неверный вид документа'
    return sRet
        
```

```python
cmd = input('Введите команду:')
if (cmd == 'p'):
    sName = input('Введите номер документа:')
    if (sName != ''):
        print(get_name(sName))
    else:
        print('Не введен номер!')
elif (cmd == 's'):
    sName = input('Введите номер документа:')
    if (sName != ''):
        print(get_shelf(sName))
    else:
        print('Не введен номер!')
elif (cmd == 'as'):
    sName = input('Введите номер новой полки:')
    if (sName != ''):
        print(add_shelf(sName))
    else:
        print('Не введен номер!')
elif (cmd == 'm'):
    doc_num = input('Введите номер документа:')
    if (is_doc(doc_num) == True):  # наличие документа
        shelf_num = input('Введите номер целевой полки:')
        if (is_shelf(shelf_num) == True):
            print(move_doc(doc_num, shelf_num))
        else:
            print('Нет такой целевой полки')
    else:
        print ('Нет такого документа')
elif (cmd == 'l'):
    get_list()
elif (cmd == 'd'):
    doc_num = input('Введите номер документа:')
    if (is_doc(doc_num) == True):  # наличие документа
        print(del_doc(doc_num))
    else:
        print ('Нет такого документа')
elif (cmd == 'a'):
    doc_num = input('Введите номер документа:')
    if (is_doc(doc_num) == False):  # нет такого документа
        shelf_num = input('Введите номер полки:')
        if (is_shelf(shelf_num) == True):
            doc_type = input('Введите тип документа:')
            p_name = input('Введите имя человека:')
            print(add_doc(doc_num, shelf_num, doc_type, p_name))
        else:
            print('Нет такой полки')
    else:
        print('Такой документ есть')
else:
    print('Неверная команда')
```
